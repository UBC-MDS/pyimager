```python
import pandas as pd
```

```python
sept= pd.read_csv("Downloads/accountactivity_sep.csv",header=None)
octo= pd.read_csv("Downloads/accountactivity_oct.csv",header=None)
nov= pd.read_csv("Downloads/accountactivity_nov.csv",header=None)
dec= pd.read_csv("Downloads/accountactivity_dec.csv",header=None)
jan= pd.read_csv("Downloads/accountactivity_jan.csv",header=None)
feb= pd.read_csv("Downloads/accountactivity_feb.csv",header=None)
```

```python
sept
```

```python
frames = [sept,octo,nov,dec,jan,feb]
df= pd.concat(frames)
```

```python
df = df.fillna(0)
expense = df[[2]]
expense['expense']=expense[2]
expense['index']=expense.index
#del expense[2]
expense
```

```python
expense

df2 = expense.head(500)
df2=df2[(df2['expense']>20) & (df2['expense']<600)]
```
